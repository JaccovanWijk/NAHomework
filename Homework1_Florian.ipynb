## Very short introduction to Matplotlib

`matplotlib` is a useful package for visualizing data using Python. Run the first cell below to plot $\sqrt{x}, x, x^2, x^3$ for $x \in [1, 10]$.

```python
x = np.linspace(1, 10, 10)  # 10 points evenly between 1 and 10.
print(x)
plt.plot(x, x**0.5, label=r"$x^{1/2}$")
plt.plot(x, x**1, label=r"$x$")
plt.plot(x, x**2, label=r"$x^2$")
plt.plot(x, x**3, label=r"$x^2$")
plt.legend()
plt.show()
```

When visualizing functions where $y$ has many different orders of magnitude, a logarithmic scale is useful:

```python
x = np.linspace(1, 10, 10)
plt.semilogy(x, x**0.5, label=r"$x^{1/2}$")
plt.semilogy(x, x**1, label=r"$x$")
plt.semilogy(x, x**2, label=r"$x^2$")
plt.semilogy(x, x**3, label=r"$x^2$")
plt.legend()
plt.show()
```

When also the $x$-axis contains many orders of magnitude, a log-log plot is most useful:

```python
x = np.logspace(1, 10, 10, base=10)  # 10 points evenly between 10^1 and 10^10.
print(x)

plt.plot(x, x**0.5, label=r"$x^{1/2}$")
plt.plot(x, x**1, label=r"$x$")
plt.plot(x, x**2, label=r"$x^2$")
plt.plot(x, x**3, label=r"$x^2$")
plt.legend()
plt.show()

plt.loglog(x, x**0.5, label=r"$x^{1/2}$")
plt.loglog(x, x**1, label=r"$x$")
plt.loglog(x, x**2, label=r"$x^2$")
plt.loglog(x, x**3, label=r"$x^2$")
plt.legend()
plt.show()
```

## (b)
Write a python function that approximates $f$ using the Taylor series. Implement a reasonable stopping criterion, so that you get an accurate result while doing more or less the minimal amount of computations. Apply your function to $x = 10^{-1}, 10^{-2}, \ldots, 10^{-16}$. 

If we compute a taylor series untill n = 5, we conserve computing power while still being accurate.
We reduce the expression for the taylor series to:
$T = \frac{(i)^{i}+1^{n}}{2(2+n)!}\cdot x^{n}$ 
For n > -2 
Putting this in code:

```python
def taylor_fun(x,n):
    sum = 0
    for i in range(1,n):
        sum = sum + ( ( (-1)**(i+1) * x**(2*i-2) ) / 2*math.factorial(i) )
    return sum

def fun(x):
    return ((1-math.cos(x))/x**2)

def new_fun(x):
    return (2*math.sin(x/2)/x**2)

def fun_run(high,low,type):
    normal_x = []
    normal_y =[]
    t_x = []
    t_y =[]
    n_x = []
    n_y = []
    if type == 'T':
        for x in np.arange(-high,low,0.001):
            T = taylor_fun(10**x,10)
            #print(f"The taylor approximation of x = 10^{x} returns {T}")
            t_x.append(10**x)
            t_y.append(T)
        return t_x,t_y
    if type == 'N':
        for x in np.arange(-high,low,0.001):
            Y = fun(10**x)
            #print(f"The normal function with x = 10^{x} returns {Y}")
            normal_x.append(10**x)
            normal_y.append(Y)
        return normal_x,normal_y
    if type == 'New':
        for x in np.arange(-high,low,0.001):
            S = new_fun(10**x)
            n_x.append(10**x)
            n_y.append(S)
        return n_x,n_y
        
# for x in np.arange(-16,0,0.1):
#     T = taylor_fun(10**x,5)
#     print(f"The taylor approximation of x = 10^{x} returns {T}")
#     t_x.append(x)
#     t_y.append(T)
```
